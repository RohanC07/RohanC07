## (5) Visualization

```python
# Top 10 best-selling games
top_games = df[['title', 'total_sales']].dropna().groupby('title').sum().nlargest(10, 'total_sales')

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(y=top_games.index, x=top_games['total_sales'], palette="viridis")
plt.xlabel("Total Sales (in millions)")
plt.ylabel("Game Title")
plt.title("Top 10 Best-Selling Video Games")
plt.show()
```
